```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(titulo, image):
    cv2.imshow(titulo, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
cats = cv2.imread("assets/fotos/cats.jpg")
show_img("Gatos", cats)

canva = np.zeros(cats.shape[:2], dtype='uint8')
```

```python
cinza = cv2.cvtColor(cats, cv2.COLOR_BGR2GRAY)
show_img("Gatos", cinza)
```

```python
cinza_hist = cv2.calcHist([cinza], [0], None, [256], [0,256])

plt.figure()
plt.title("Histograma")
plt.xlabel("Intervalos/Setores")
plt.ylabel("Número de Pixels")
plt.plot(cinza_hist)
plt.xlim([0,256])
plt.show()
```

```python


plt.figure()
plt.title("Histograma de Cores")
plt.xlabel("Intervalos/Setores")
plt.ylabel("Número de Pixels")
colors = ['b', 'g', 'r']

for channel, cor in enumerate(colors):
    hist = cv2.calcHist([cats], [channel], None, [256], [0,256])
    plt.plot(hist, color = cor)
    plt.xlim([0,256])

plt.show()
```
